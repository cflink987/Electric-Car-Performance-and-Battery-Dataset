```python
#This series of code allowed me to seperate my battery sizes into 3 groups Small, Medium, and Large by usung IQR. 
#We see that Moderate is the largest due to having 50% of the datasets batteries fall in it by design.

df["battery_capacity_kWh"] = pd.to_numeric(df["battery_capacity_kWh"], errors="coerce")

q1, q3 = df["battery_capacity_kWh"].quantile([0.25, 0.75])

df["battery_size_3"] = np.where(df["battery_capacity_kWh"] < q1, "Small",
                         np.where(df["battery_capacity_kWh"] > q3, "Large", "Moderate"))
df["battery_size_3"] = pd.Categorical(df["battery_size_3"],
                                      categories=["Small", "Moderate", "Large"], ordered=True)

df["battery_size_3"].value_counts(sort=False)
```
